```python
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
# Sample legitimate transactions to balance the dataset
normal_sample = normal.sample(n=492)

# Concatenate legitimate_sample and fraudulent transactions to create a new dataset
new_df = pd.concat([normal_sample, fraud.sample(n=492)], axis=0)

# Calculate and visualize the class distribution of the new dataset
class_counts = new_df['Class'].value_counts()
print(class_counts)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
colors = ['#4C72B0', '#55A868']  
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Class Distribution (Balanced)')
plt.axis('equal')
```

```python
X = new_df.drop(columns=['Class'])
Y = new_df['Class']
# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

models = [RandomForestClassifier(), LogisticRegression(), XGBClassifier(), AdaBoostClassifier()]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a list of models defined as 'models'
for i, m in enumerate(models):
    print(m)
    m.fit(X_train, Y_train)
    print(m, m.score(X_train, Y_train))
    
    y_pred = m.predict(X_test)
    print(classification_report(Y_test, y_pred))
    
    cm = pd.crosstab(Y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
    
    f, ax = plt.subplots(figsize=(5,5))
    sns.heatmap(cm, annot=True, linewidth=0.7, linecolor='black', fmt='g', ax=ax, cmap="BuPu")
    plt.title(f'{m} - Confusion Matrix')
    plt.xlabel('Y predict')
    plt.ylabel('Y test')
    plt.show()

 

```
