```python
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data= df, x= 'Class')
plt.title('Distribution of normal and fraud transactions')
plt.show()
```

```python
normal_sample = normal.sample(n=492)
nv_df = pd.concat([normal_sample, fraud.sample(n=492)], axis = 0)

class_counts = nv_df['Class'].value_counts()
```

```python
print(class_counts)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
colors = ['#4C72B0', '#55A868']
plt.pie(class_counts, labels = class_counts.index, autopct = '%1.1f%%', startangle = 140, colors = colors)
plt.title('Class Distribution')
plt.axis ('equal')
```

```python
X = nv_df.drop(columns = ['Class'])
Y = nv_df['Class']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
models = [RandomForestClassifier(), LogisticRegression(), AdaBoostClassifier(), XGBClassifier()]
```

```python
for i, m in enumerate(models):
    print(m)
    m.fit(X_train, Y_train)
    print(m, m.score(X_train, Y_train))

    y_pred = m.predict(X_test)
    print(classification_report(Y_test,y_pred))

    cm = pd.crosstab(Y_test, y_pred, rownames =['Actual'], colnames = ['Predicted'])
    f,ax = plt.subplots(figsize=(5,5))
    sns.heatmap(cm, annot = True, linewidth = 0.7, linecolor = 'black', fmt = 'g', ax =ax, cmap = 'BuPu')
    plt.title(f'{m} - confusion matrix')
    plt.xlabel('Y predict')
    plt.ylabel('Y test')
    plt.show()
```

LogisticRegression() is the best model
